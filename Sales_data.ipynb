```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import tha excel file
df = pd.read_excel('/content/sales (2).xlsx')
```

```python
df.head()
```

```python
# to get the information baout dataset
df.info()
```

```python
# To get the number of missing values in each column
df.isnull().sum()
```

```python
# drop a column
df.drop('id', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Draw a bar chart
df['itemName'].value_counts().plot(kind='bar')
```

```python
df['batchNo'].value_counts().plot(kind='bar')
```

```python
# want to get sales quntities in month wise
df['orderDate'].value_counts()

# show this in bar chart
df['orderDate'].value_counts().plot(kind='bar')
```

```python
# from the order data column, I want to get the number of the month
df['orderDate'] = pd.to_datetime(df['orderDate'])

# Then add to new column name month number
df['monthNumber'] = df['orderDate'].dt.month
```

```python
# check head
df.head()
```

```python
# according to the monthNumber get the quantities
df.groupby('monthNumber')['quantity'].sum()

# show in a bar chart
df.groupby('monthNumber')['quantity'].sum().plot(kind='bar')
```

```python
# show items sales quantity in line chart
df.groupby('itemName')['quantity'].sum().plot(kind='line')
```

```python
# show sales quntity in a pie chart by itemName wise
df.groupby('itemName')['quantity'].sum().plot(kind='pie')

# need to show the precentage also
df.groupby('itemName')['quantity'].sum().plot(kind='pie', autopct='%1.1f%%')
```
